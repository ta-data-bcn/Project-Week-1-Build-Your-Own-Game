```python
import random
import pandas as pd

#myDataFrame useful to get a word and play the hangame.
pokemon_df = pd.read_csv("/Users/Espadas/pokemon_data.csv", encoding='utf-8')
list_pokemons=[value.upper() for value in pokemon_df['Name']]
list_type_pokemons=[value.upper() for value in pokemon_df['Type 1']]

movies_df=pd.read_csv("/Users/Espadas/movies.csv", error_bad_lines=False) 
list_movies=[value.upper() for value in movies_df['Film']]
list_type_movies=[value.upper() for value in movies_df['Genre']]

selection_topics = {'POKEMON':{'WORD':list_pokemons, 'HINT':list_type_pokemons}, 'MOVIES':{'WORD':list_movies,'HINT':list_type_movies}}


def ask_category(): #Player choose the topic thanks to dataframe   
    print("Let's start with Hangman game!\n")
    print(display_hangman(0))
    
    while True:
        try:
            print(f'\nToday we have theses topics: {list(selection_topics.keys())}')
            player_category_desired = input(str("Wich topic you wanna play? \n")).upper()
            if player_category_desired in list(selection_topics.keys()):
                break
        except ValueError:
            print("\nUps... make sure to write a string! Please, write the topic again.")
    return player_category_desired

def get_word(player_category_desired): #choosing the word in random way
    word = random.choice(selection_topics[player_category_desired]['WORD'])
    word = word.strip(",.?!*")
    return word.upper()

def get_hint(word, player_category_desired): #choosing the hint according the index of the lists's word
    hint = selection_topics[player_category_desired]['WORD'].index(word)
    word_hint = selection_topics[player_category_desired]['HINT'][hint]
    return word_hint.upper()

def presentation_tries(tries,word_completion,hint_completion):  #i call this function as the interface of the game where player will play
    print(display_hangman(tries))
    print(word_completion)
    print(hint_completion)
    print("\n")

def get_word_completion(word): #making the '_' according the length's word
    word_completion=''
    for i in word:
        if i != " ":
            word_completion = word_completion+"_" #if there's no space, add '_' because means there's a letter to guess by player
        else:
            word_completion = word_completion+i #we want spaces and not '_' because if there're '_' meanse there's a letter to guesses and is not a letter is a space, so we avoid the spaces
    return word_completion
    
def play(word, hint): #let's start to write the rules the game, first declaring variables
    word_completion = get_word_completion(word)
    hint_completion = "Here's the word's hint: " + hint
    guessed = False
    guessed_letters = []
    guessed_words = []
    tries = 7 #take an eye there're 8 hanggames graphics, so 8 situations differents, so 8 tries
    presentation_tries(tries, word_completion, hint_completion)#we show the interface game for the player
    while not guessed and tries > 0: #this will work if the player not discover the word (win) and the tries keeps upper 0(loose)
        guess = input(str("Please guess a letter or word: ")).upper() #all the strings are upper cases according to dataframe
        if len(guess) == 1 and guess.isalpha():
            if guess in guessed_letters: #First of all, we check if that letter is guessed before (although would be the first time)
                print("You already guessed the letter", guess)
            elif guess not in word: #if the before condition is not right, here we see if there's a failed letter by player and append to list of letters guessed and substract tries
                print(guess, "is not in the word.")
                tries -= 1
                guessed_letters.append(guess)
            else:
                print("Good job,", guess, "is in the word!")
                guessed_letters.append(guess)
                word_as_list = list(word_completion)#we convert the string word_completion to list because we'll replace every string correct between '_'
                indices = [i for i, letter in enumerate(word) if letter == guess] #we enumerate the word (only if the letter guessed) to get index letter guessed form the word
                for index in indices:
                    word_as_list[index] = guess #the list word_completion that we made before, their '_' will be replaced for the letter guessed by player(or letters, imagine there's a multiple letters'k' in the word)
                word_completion = "".join(word_as_list) #we get again the string the word_completion, but this time with the letter(s) guessed by player instead of '_'
                if "_" not in word_completion: #if there is not any of '_', then we finish and player guessed the whole word
                    guessed = True
        elif len(guess) == len(word) and guess.isalpha(): #if the player try to write the whole word instead of letter by letter...
            if guess in guessed_words: #check if the word guess is in the list of wrong words written (although would be the first time)
                print("You already guessed the word", guess)
            elif guess != word: #if the world is new but is wrong... we append into the list of wrong words
                print(guess, "is not the word.")
                tries -= 1
                guessed_words.append(guess)
            else: #we win when the before condition are not true
                guessed = True
                word_completion = word
        else:
            print("Not a valid guess.")
        presentation_tries(tries, word_completion, hint_completion) # when finish the conditions, we'll show the hangman updated. This is the last sentence (and we always go on here) for the loop while
    if guessed: #look up that in the moment we won, the v ariable guessed is True, then out of the loop while we start this condition
        print("Congrats, you guessed the word! You win!")
    else:
        print("Sorry, you ran out of tries. The word was " + word + ". Maybe next time!")


def display_hangman(tries):
    stages = [  
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / \ HANGMAN!
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / \ ONE MORE TRY!
                   -
                """,
                
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |     / 
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \|/
                   |      |
                   |      
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |     \|
                   |      |
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      |
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      O
                   |      |
                   |      
                   |     
                   -
                """,
                """
                   --------
                   |      |
                   |      
                   |    
                   |      
                   |     
                   -
                """
    ]
    return stages[tries]

def play_again_or_not():
    try:
        while True:
            continue_game = input("Play Again? (Y/N) \n").upper()
            if continue_game == 'Y':
                return True
                break
            elif continue_game == 'N':
                return False
                break
    except:
        pass
                
def main():
    play_on = True
    while play_on == True:
        player_category_desired = ask_category()
        word = get_word(player_category_desired)
        hint = get_hint(word,player_category_desired)
        play(word,hint)
        play_on=play_again_or_not()
    print("\nSee you the next time!")


if __name__ == "__main__":
    main()
```
