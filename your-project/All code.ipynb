```python
#Ask for the level they want to play
levels=["easy", "medium", "hard"]
def ask_level():
    choice=input("Pick your level: easy, medium or hard? ").lower()
    while choice not in levels:
        print("Please, choose one of the options: easy, medium or hard")
        choice=input("\nPick your level: easy, medium or hard? ").lower()
    return choice
level=ask_level()

#Set range of numbers and number of tries according to the level chosen
numbers=[]
if level=="easy":
    for i in range(1,11):
        numbers.append(i)
    tries=5
    print("Great! You'll have to pick a number between 1 and 10. You have 5 turns to get it right. Let's play!\n")
elif level=="medium":
    for i in range(1,101):
        numbers.append(i)
    tries=7
    print("Great! You'll have to pick a number between 1 and 100. You have 7 turns to get it right. Let's play!\n")
else:
    for i in range(1,1001):
        numbers.append(i)
    tries=10
    print("Great! You'll have to pick a number between 1 and 1000. You have 10 turns to get it right. Let's play!\n")


#Generate random correct value
import random
correct_number=random.choice(numbers)


#Define function for user's guessed number
ask_number_text="Pick a number between " + str(min(numbers)) + " and " + str(max(numbers)) + ": "
str_numbers=[]
for i in range(0,len(numbers)):
    str_numbers.append(str(i))
    
def ask_number():
    choice1=input(ask_number_text)
    while choice1 not in str_numbers:
        print("Please, pick a number within the range\n")
        choice1=input(ask_number_text)
    return int(choice1)

#Run game
tries_count=0
for i in range(0, tries):
    guess_number=ask_number()
    if guess_number==correct_number:
        print("Congratulations! You guessed the number!")
        break
    elif guess_number<correct_number and tries_count<tries-1:
        tries_count+=1
        print("You are too low, try again! Tries left:", tries-tries_count, "\n")
        continue
    elif guess_number>correct_number and tries_count<tries-1:
        tries_count+=1
        print("You are too high, try again! Tries left:", tries-tries_count, "\n")
        continue
    else:
        print("You don't have any tries left, you lose the game")
```
