```python
mastermind_text = """

%%%%%%%%%%%%%%% WELCOME TO THE MASTERMIND GAME %%%%%%%%%%%%%%%

The purpose of the game is to guess which balls the CPU has chosen and in what order.

The CPU will choose 4 of the 8 possible balls. The balls are:

- Red ball
- Green ball
- Blue ball
- Yellow ball
- Brown ball
- Orange ball
- Black ball
- White ball

When it comes your turn to chose, keep in mind that the order in which you choose the balls is the key to win the game.

At the end of each round you will know how many of the balls you have chosen:
     1. They have the right color and are in the right position.
     2. They just have the right color, but they are not in their position.

The difficulty of the game is that you will not know which colors you have guessed and, in case you have guessed one or some of them, which ones are in the correct positions.

To achieve this, you have a maximum of 9 rounds. If you don't get it right at the end of the ninth round...you lose.

Do you find it too difficult? I will give you another clue: The CPU can only choose one ball of each color. In other words, there are no duplicate balls.

Good luck!

_______________________________________________________________________________________________________________________________________________________________________________

"""
print(mastermind_text)

n_balls = 4
max_rounds = 9
ball_colours = ['red', 'green', 'blue', 'yellow', 'brown', 'orange', 'black', 'white']

import random
def cpu_choice():
    random.shuffle(ball_colours)
    ball_positions = ball_colours[:n_balls]
    return(ball_positions)
cpu_balls = cpu_choice()

def player_choice():
    player_balls = []
    while True:
        ball1 = input(f'Make your first choice. Remember, the colour of the balls are red, green, blue, yellow, brown, orange, black and white: ')
        if ball1 in ball_colours:
            player_balls.append(ball1)
            print(f'You chose the {ball1} ball in slot 1.\n')
            break
        elif ball1 not in ball_colours:
            print('You entered a wrong input. Please check your spelling and try again.\n')
            continue
    while True:
        ball2 = input(f'Make your second choice. Remember, the colour of the balls are red, green, blue, yellow, brown, orange, black and white: ')
        if ball2 in ball_colours:
            player_balls.append(ball2)
            print(f'You chose the {ball2} ball in slot 2.\n')
            break
        elif ball2 not in ball_colours:
            print('You entered a wrong input. Please check your spelling and try again.\n')
            continue
    while True:
        ball3 = input(f'Make your third choice. Remember, the colour of the balls are red, green, blue, yellow, brown, orange, black and white: ')
        if ball3 in ball_colours:
            player_balls.append(ball3)
            print(f'You chose the {ball3} ball in slot 3.\n')
            break
        elif ball3 not in ball_colours:
            print('You entered a wrong input. Please check your spelling and try again.\n')
            continue
    while True:
        ball4 = input(f'Make your fourth choice. Remember, the colour of the balls are red, green, blue, yellow, brown, orange, black and white: ')
        if ball4 in ball_colours:
            player_balls.append(ball4)
            print(f'You chose the {ball4} ball in slot 4.\n')
            break
        elif ball3 not in ball_colours:
            print('You entered a wrong input. Please check your spelling and try again.\n')
            continue
    return player_balls

rounds_played = 0
while rounds_played != max_rounds:
    rounds_played += 1
    print(f"""
ROUND Nº: {rounds_played} OF {max_rounds}
––––––––––––––––
""")
    player_balls = player_choice()
    balls_colour_check = []
    balls_position_check = 0
    for colours in player_balls:
        if colours in cpu_balls:
            balls_colour_check.append(colours)
    for player_positions, cpu_positions in zip(player_balls, cpu_balls):
        if player_positions == cpu_positions:
            balls_position_check += 1
    if balls_position_check == 4:
        print(f"""
The order of the balls you have chosen is: {' / '.join(player_balls)}
And the order of the balls CPU has chosen is:  {' / '.join(cpu_balls)}
CONGRATULATIONS! YOU WON!""")
        break
    if rounds_played == max_rounds:
        print(f"""
You have run out of rounds.
The order of the balls CPU has chosen is: {' / '.join(cpu_balls)}
I'M SORRY. YOU LOSE!

""")
        break
    print('The order of the balls you have chosen is:', ' / '.join(player_balls))
    print(f'You guessed the color of: {len(balls_colour_check)} of 4 balls.')
    print(f"""You guessed the position of: {balls_position_check} of 4 balls.

    """)
```
