# Black Jack
pack = 52 cards
objective = get score 21 or as close as possible without going over 21
#### card values
ace = 1 or 11
face cards = 10
any other card is its pip value
#### betting
first step = you make the bet up to 500€
#### the game
round 1
player gets a card and dealer gets a card
round 2
player gets a card and dealer gets a card faced down

if player scores 21 then he/she gets 2.5 times the bet as a reward

if dealer first card is not a face card or ace then card 2 remain hidden
if dealer first card is a face card or ace then we unveil the second card
if dealer scores 21 then the player loses the bet
if both have 21 (tie) then the bet is in stand-off

the play
player knows his 2 cards values then he decides:
    stand = not ask for another card
    hit = ask for another card
the aim is to get as close to 21 as possible
but if it goes > 21 then it's a burst which means he loses the game

dealer play, if we reach this point, now we show the result of card 2 
if total is >= 17 <= 21 then stand and compare to player cards
if total is < 17 then get extra card until total is >= 17 <= 21
