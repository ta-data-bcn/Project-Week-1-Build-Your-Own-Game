**Original version:**
It puts the 2 Players in one territory/district each one. The objective of the game is to attack and conquer the territory where the other player is.
The game finishes when one of the two players have conquered all territories.

**Noob Ironhacker version:**
We'll play a simple, but functional demo composed by:

- 2 Players (one soldier each one)
- 2 districts
- 2 dices
- 1 Winner

"*Will you be the next to conquer the city?*"

```python
import random
import time

district_list = ['POBLENOU', 'EIXAMPLE']
wannaplay=True

print("Hi Ironhackers, did you survive the first week? Cool!! Honestly, I didn't even believe in myself. But now, let's play Risk game to have fun.")
print("We'll start with 2 districts: ", district_list)
time.sleep(8)

while wannaplay:
    while True:
        try:
            district_a = str(input("Player 1, in which district do you want to start? >>> ")).upper()
            if district_a in district_list:
                while True:
                    district_b = str(input("Player 2, in which district do you want to start? >>> ")).upper()
                    if district_b in district_list:
                        if district_a != district_b:
                            break
                        else:
                            print("This district is already taken. Hint: try the other one.")
                    else:
                        print(f"The district {district_b} doesn't exist in this game. Try choosing it again.")
                        continue
                break
            else:
                print(f"The district {district_a} doesn't exist in this game. Please, repeat again! ")
                continue
        except:
            pass
    print("Let's start the Warcelona!")
    time.sleep(5)

    #Second, let's roll a dice:

    def dice_1():
        dice1 = random.randint(1,6)
        return dice1
    

    def dice_2():
        dice2 = random.randint(1,6)
        return dice2
    
    while True:
        dice1=dice_1()
        dice2=dice_2()
        if dice1 != dice2:
            break
        print("The Player1 rolled the number...", dice1)
        print("The Player2 rolled the number...", dice2)
        time.sleep(2)
        print("Tie!!! Let's roll again.")
        time.sleep(2)

    if dice1 > dice2:
        print("The Player1 rolled the number...", dice1)
        print("The Player2 rolled the number...", dice2)
        time.sleep(2)
        print("Player1 conquered the Player2 district and WON the game.")
 
    else:
        print("The Player1 rolled the number...", dice1)
        print("The Player2 rolled the number...", dice2)
        time.sleep(2)
        print("Player2 conquered the Player1 district and WON the game.")
        
    again=str(input("Do you want to play again? Type yes or no: "))
    if again == "no":
        wannaplay = False
        
print( 'Thank you for playing Risk!!!!')
```
